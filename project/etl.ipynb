```python
from feature_engineering.query_data_warehouse import query_data_warehouse
from nosql_db import NoSqlDB

db_config = {
    "mongo_uri": "mongodb://localhost:27017/",
    "db_name": "nosql_db",
    "collection_name": "ros2_doc",
}
db = NoSqlDB(db_config)
all_documents = query_data_warehouse(db, with_clearml=False)
print(f"Total documents: {len(all_documents)}")
print('-' * 20)
for document in all_documents:
    print('document type: ', document['doc_type'])
    print('document title: ', document['title'])
    print('document subtitle: ', document['subtitle'])
    print('document url: ', document['url'])
    print('-' * 20)
```
